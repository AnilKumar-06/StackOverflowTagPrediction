# Preprocessing

1.Simple 0.5M data points.

2. Seprate out code-snippets from body.

3. Remove Special characters from Question title and description (not in code).

4. Removes HTML Tags.

5. Remove stop words (Except 'C').

6. Convert all the characters into small letters.

7. Use SnowballStemmer to stem the words


```python
import nltk
nltk.download('stopwords')
```

```python
def striphtml(data):
    cleaner = re.compile('<.*?>')
    cleantxt = re.sub(cleaner, ' ', str(data))
    return cleantxt

stop_words = set(stopwords.words('english'))
stemmer = SnowballStemmer("english")
```

```python
def create_connection(db_file):
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)

    return None
```

```python
def create_table(conn, create_table_sql):
    try:
        c = conn.cursor()
        c.execute(create_table_sql)
    except Error as e:
        print(e)
```

```python
def checkTableExists(dbcon):
    cursr = dbcon.cursor()
    sql = "select name from sqlite_master where type = 'table'"
    table_names = cursr.execute(sql)
    print("Tables in the database: ")
    tables = table_names.fetchall()
    print(tables[0][0])
    return (len(tables))
```

```python
def create_database_table(database, query):
    conn = create_connection(database)
    if conn is not None:
        create_table(conn, query)
        checkTableExists(conn)
    else:
        print("Error! can't create the database connection.")
    conn.close()

sql_create_table = """CREATE TABLE IF NOT EXISTS QuestionsProcessed (question text NOT NULL, code text, tags text, 
  words_pre integer, words_post integer, is_code integer);"""
create_database_table("Processed.db", sql_create_table)
```

```python
start = dt.datetime.now()
preprocess_data = []
reader.fetchone()
ques_with_code = 0
len_pre = 0
len_post = 0
ques_processed = 0

for row in reader:
    is_code = 0
    title, ques, tags = row[0], row[1], row[2]

    if '<code>' in ques:
        ques_with_code += 1
        is_code = 1
    x = len(ques) + len(title)
    len_pre += x

    code = str(re.findall(r'<code>(.*?)</code>', ques, flags = re.DOTALL))
    
    ques=re.sub('<code>(.*?)</code>', '', ques, flags=re.MULTILINE|re.DOTALL)
    ques = striphtml(ques.encode('utf-8'))

    title = title.encode('utf-8')

    ques = str(title) + " " +str(ques)
    ques = re.sub(r'[^A-Za-z]+', ' ', ques)
    words = word_tokenize(str(ques.lower()))

    #Removing all single letter and and stopwords from question except for the letter 'c'
    ques = ' '.join(str(stemmer.stem(j)) for j in words if j not in stop_words and (len(j) != 1 or j == 'c'))
    len_post += len(ques)

    tup = (ques, code, tags, x, len(ques), is_code)
    ques_processed += 1
    #insrt_sql = 'INSERT INTO QuestionsProcessed(question, code, tags, words_pre, words_post, is_code) values (?, ?, ?, ?, ?, ?)';
    sql_writer.execute('INSERT INTO QuestionsProcessed(question, code, tags, words_pre, words_post, is_code) values (?, ?, ?, ?, ?, ?)', tup)

    if (ques_processed % 100000 == 0):
        print("Number of questions = ", ques_processed)

no_dup_avg_len_pre = (len_pre * 1.0)/ques_processed
no_dup_avg_len_post = (len_post * 1.0)/ ques_processed

print("Avg length of Question(Title + Body) before processing: ", no_dup_avg_len_pre)
print("Avg length of Question(Title + Body) before processing: ", no_dup_avg_len_post)
print("Percent of questions containing code : ", ((ques_with_code * 100.0)/ques_processed))
print("Total Time taken = ", dt.datetime.now() - start)
```

```python
vect = CountVectorizer(tokenizer = lambda x: x.split(), binary = True)
multilabel_y = vect.fit_transform(preprocess_data['tags'])
```

We will sample the number of tags instead considering all of them (due to limitation of computing power)

```python
def tags_to_choose(n):
    t = multilabel_y.sum(axis = 0).tolist()[0]
    sorted_tags_i = sorted(range(len(t)), key = lambda i : t[i], reverse = True)
    multilabel_yn = multilabel_y[:, sorted_tags_i[:n]]
    return multilabel_yn
```

```python
def questions_explained_fn(n):
    multilabel_yn = tags_to_choose(n)
    x = multilabel_yn.sum(axis = 1)
    return (np.count_nonzero(x == 0))
```

```python
ques_explained = []
total_tags = multilabel_y.shape[0]
total_ques = preprocess_data.shape[0]
for i in range(500, total_tags, 100):
    ques_explained.append(np.round(((total_ques - questions_explained_fn(i))/total_ques)*100, 3))
```

# Preprocessing Questions

Seprate code from Body

Remove special Characters from question, Title and Descriptions(not from code)

Gie more weightage to title: Add title three times to the question

```python
start = dt.datetime.now()
preprocess_data = []
reader.fetchone()
ques_with_code = 0
len_pre = 0
len_post = 0
ques_processed = 0

for row in reader:
    is_code = 0
    title, ques, tags = row[0], row[1], row[2]

    if '<code>' in ques:
        ques_with_code += 1
        is_code = 1
    x = len(ques) + len(title)
    len_pre += x

    code = str(re.findall(r'<code>(.*?)</code>', ques, flags = re.DOTALL))
    ques=re.sub('<code>(.*?)</code>', '', ques, flags=re.MULTILINE|re.DOTALL)

    ques = striphtml(ques.encode('utf-8'))

    title = title.encode('utf-8')

    ques = str(title) + " " +str(ques)
    ques = re.sub(r'[^A-Za-z0-9#+.\-]+', ' ', ques)
    words = word_tokenize(str(ques.lower()))

    #Removing all single letter and and stopwords from question except for the letter 'c'
    ques = ' '.join(str(stemmer.stem(j)) for j in words if j not in stop_words and (len(j) != 1 or j == 'c'))
    len_post += len(ques)

    tup = (ques, code, tags, x, len(ques), is_code)
    ques_processed += 1
    insrt_sql = 'INSERT INTO QuestionsProcessed(question, code, tags, words_pre, words_post, is_code) values (?, ?, ?, ?, ?, ?)';
    writer.execute(insrt_sql, tup)

    if (ques_processed % 100000 == 0):
        print("Number of questions = ", ques_processed)

no_dup_avg_len_pre = (len_pre * 1.0)/ques_processed
no_dup_avg_len_post = (len_post * 1.0)/ ques_processed

print("Avg length of Question(Title + Body) before processing: ", no_dup_avg_len_pre)
print("Avg length of Question(Title + Body) before processing: ", no_dup_avg_len_post)
print("Percent of questions containing code : ", ((ques_with_code * 100.0)/ques_processed))
print("Total Time taken = ", dt.datetime.now() - start)
```

```python
from sklearn import metrics
def model_Perfoemance(y_pred, y_test):

    print("Accuracy = ",metrics.accuracy_score(y_test, pred))
    print("Hamming Loss = ", metrics.hamming_loss(y_test, pred))

    micro_precision = precision_score(y_test, pred, average = 'micro')
    micro_recall = recall_score(y_test, pred, average = 'micro')
    micro_f1 = f1_score(y_test, pred, average = 'micro')

    print("Micro-average quality numbers")
    print("Micro_Precision: {:.4f}, Micro_Recall : {:.4f}, Micro_F1-Measure : {:.4f}".format(micro_precision, micro_recall, micro_f1))

    macro_precision = precision_score(y_test, pred, average = 'macro')
    macro_recall = recall_score(y_test, pred, average = 'macro')
    macro_f1 = f1_score(y_test, pred, average = 'macro')

    print("Macro-average quality numbers")
    print("Macro_Precision : {:.4f}, Macro_Recall : {:.4f}, Macro_F1-Measure : {:.4f}".format(macro_precision, macro_recall, macro_f1))
```
